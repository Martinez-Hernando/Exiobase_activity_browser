## Technosphere matrix (Z value)

Values

```python
tec_data=exio3318.Z.values.flatten()
tec_data=np.nan_to_num(tec_data)
print(type(tec_data))
```

Index

```python
tec_index=itertools.product(range(n), range(n))
tec_index=list(tec_index)
tec_index=np.array(tec_index, dtype=bwp.INDICES_DTYPE)

print(len(tec_index))
tec_index
```

## Biosphere matrix

Values

```python
b_data=np.array([exio3318.emissions.F.iloc[0,:]]).T
b_data=b_data.squeeze()
b_data=np.nan_to_num(b_data)
print(b_data)
```

Index

```python
b_index=itertools.product(emissions, nick)
b_index=list(b_index)
b_index=np.array([b_index],dtype=bwp.INDICES_DTYPE).T
b_index=b_index.squeeze()
print(b_index.shape)
b_index
```

## Method

```python
c_data=np.array([1])
print(c_data)
c_index=np.array([(CO2, CO2)] ,dtype=bwp.INDICES_DTYPE)
print(c_index.shape)
```

# Creation of datapackage

```python
dp_static = bwp.create_datapackage()
dp_static.add_persistent_vector(
    matrix='technosphere_matrix',
    indices_array=tec_index,
    data_array=tec_data,
)
dp_static.add_persistent_vector(
    matrix='biosphere_matrix',
    indices_array=b_index,
    data_array=b_data,
)
dp_static.add_persistent_vector(
    matrix='characterization_matrix',
    indices_array=c_index,
    data_array=c_data,
)
```

```python
dp_static.finalize_serialization
```

# LCA

```python
lca = bc.LCA(
    demand={225: exio3318.x.iloc[224, 0]},  
    data_objs=[dp_static],
)
lca.lci()
lca.lcia()
lca.score
print(lca.score)
```
